```python
def fib_recursion(n):
    n_fibonacci = 0
    if n == 0:
        return n_fibonacci
    if n == 1:
        n_fibonacci+=1
        return n_fibonacci
    else:
        n_fibonacci=fib_recursion(n-1)+fib_recursion(n-2)
        return n_fibonacci
```

```python
import time
import matplotlib.pylab as plt
times = []
iteration = []
for i in range(40):
    iteration.append(i+1)
    init=time.time()
    fib_recursion(i+1)
    difference_time=time.time()-init
    times.append(difference_time)
plt.plot(iteration, times)
plt.title("Fibonacci recursivo")
plt.xlabel("Iteraciones")
plt.ylabel("Tiempo(s)")
plt.show()
```

```python
def fib(n):
    a=1
    b=0
    for i in range(n):
        n = a
        a = b
        b += n
    return b
```

```python
times = []
iteration = []
for i in range(100):
    iteration.append(i+1)
    init=time.time()
    fib(i+1)
    difference_time=time.time()-init
    times.append(difference_time)
plt.plot(iteration, times)
plt.title("Fibonacci no recursivo")
plt.xlabel("Iteraciones")
plt.ylabel("Tiempo(s)")
plt.show()
```

La principal diferencia entre los dos gráficos es que el del método no recursivo muestra tiempos mucho más bajos y aunque también parece acumulativo no crece de manera exponencial como el del gráfico del método recursivo, esto debido a que el método recursivo se llama a si mismo cada vez más veces entre más grande sea el termino de la serie de Fibonacci que se requiera. Por lo cual se puede concluir que el método recursivo es completamente ineficiente para obtener terminos grandes (>30) de la serie de Fibonacci.

```python
cociente = []
iteration = []
for i in range(100):
    iteration.append(i+1)
    f=fib(i+1)
    f1=fib(i+2)
    c=f1/f
    cociente.append(c)
plt.plot(iteration, cociente)
plt.title("Cocientes serie de Fibonacci")
plt.xlabel("Iteraciones")
plt.ylabel("Cociente")
plt.show()
```

```python
import math
def euler(n):
    h = 1/n
    x = 0
    X = []
    y = 1
    e = []
    for i in range(n):
        X.append(x)
        z = 1+0.5*math.exp(-4*x)-0.5*math.exp(-2*x)
        e.append(abs(y-z)/z)
        y += h * (2-math.exp(-4*x)-2*y)
        x += h
    X.append(x)
    z = 1+0.5*math.exp(-4*x)-0.5*math.exp(-2*x)
    e.append(abs(y-z)/z)
    mean = sum(e)/(n+1)
    plt.plot(X, e)
    plt.title("Error método de Euler")
    plt.xlabel("t")
    plt.ylabel("Error")
    plt.show()
    return "%.6f" % mean
a= euler(17)
print (a)
```

La principal fuente de error es el salto h, entre más grande sea este número mayor va a ser el error relativo, puesto que entre más grande sea el salto más se va a alejar la linea tangente del valor verdadero.

```python
def listOfLists(l):
    for i in range(len(l)):
        for j in range(len(l)):
            if l[j]>l[i]:
                n=l[i]
                l[i]=l[j]
                l[j]=n
    ll = []
    for i in l:
        n = []
        for j in range(i):
            n.append(i)
        ll.append(n)
    return ll
l = [1,3,2,8,6]
print(listOfLists(l))
```
