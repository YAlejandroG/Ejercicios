```python
def randomGeneration(a, c, m, r):
    randomSequence = [r]
    for i in range(m-1):
        r1 = (a*r+c)%m
        r = r1
        randomSequence.append(r)
    return randomSequence
r = randomGeneration(57, 1, 256, 10)
t = len(r)
print (r)
print ('El periodo de la secuencia generada es: ', t)
```

```python
import math
def testUniformity(k, N, r):
    r1=[]
    for j in r:
        n = j**k-1/(k+1)
        r1.append(n)
    c = math.sqrt(N)*abs(sum(r1)/N)
    return c
```

```python
k=[1, 3, 7]
N=[100, 10000, 100000]
r3=np.random.random(1000)
for i in k:
    for j in N:
        a = testUniformity(i, j, r3)
        print ("Para k=", i, " y N=", j, ": ", a)
print("Por lo cual el metodo random.random() genera una distribucin uniforme")
```
