```python
import math
import numpy as np
import matplotlib.pyplot as plt

def caminata2D(n=1000):
    deltaX = (np.random.random(n)-0.5)*2
    deltaY = (np.random.random(n)-0.5)*2
    L = np.sqrt(deltaX**2 + deltaY**2)
    deltaX = deltaX/L
    deltaY = deltaY/L
    x=np.cumsum(deltaX)
    y=np.cumsum(deltaY)
    r2 = x[-1]**2 + y[-1]**2
    return {'x':x, 'y':y, 'deltaX':deltaX, 'deltaY':deltaY, 'r2':r2}
```

```python
def caminatas2D(n=1000):
    k=int(math.sqrt(n))
    r=[]
    m={}
    for i in range(k):
        m[i] = caminata2D(n)
        r.append(m[i]['r2'])
    rProm=sum(r)/k
    return m, rProm
```

```python
def plotCaminatas2D(n=1000):
    m, r = caminatas2D(n)
    plt.figure()
    for i in m.keys():
        plt.plot(m[i]['x'], m[i]['y'], linewidth=0.4)
    plt.title("Caminatas aleatorias 2D")
    plt.savefig("Caminatas2D.png")
    return r

r1 = plotCaminatas2D()
print('R2 promedio de caminatas aleatorias 2D: ', r1)
```

```python
def caminata3D(n=1000):
    deltaX = (np.random.random(n)-0.5)*2
    deltaY = (np.random.random(n)-0.5)*2
    deltaZ = (np.random.random(n)-0.5)*2
    L = np.sqrt(deltaX**2 + deltaY**2 + deltaZ**2)
    deltaX = deltaX/L
    deltaY = deltaY/L
    deltaZ = deltaZ/L
    x=np.cumsum(deltaX)
    y=np.cumsum(deltaY)
    z=np.cumsum(deltaZ)
    r2 = x[-1]**2 + y[-1]**2 + z[-1]**2
    return {'x':x, 'y':y, 'z':z, 'deltaX':deltaX, 'deltaY':deltaY, 'deltaZ':deltaZ, 'r2':r2}
```

```python
def caminatas3D(n=1000):
    k=int(math.sqrt(n))
    r=[]
    m={}
    for i in range(k):
        m[i] = caminata3D(n)
        r.append(m[i]['r2'])
    rProm=sum(r)/k
    return m, rProm
```

```python
from mpl_toolkits.mplot3d import Axes3D
def plotCaminatas3D(n=1000):
    m, r = caminatas3D(n)
    fig = plt.figure()
    ax = Axes3D(fig)
    for i in m.keys():
        ax.plot(m[i]['x'], m[i]['y'], m[i]['z'], linewidth=0.4)
    plt.title("Caminatas aleatorias 3D")
    plt.savefig("Caminatas3D.png")
    return r
    
r2 = plotCaminatas3D()
print('R2 promedio de caminatas aleatorias 3D: ', r2)
```

```python
def decay(N, l=0.2*10**6):
    t0 = 0
    t = []
    n = []
    dn = []
    while N>0:
        dN = 0
        for i in range(N):
            if np.random.random()*10**6<l:
                dN+=1
        t0+=1
        t.append(t0)
        N-=dN
        if N>0 and dN>0:
            n.append(math.log(N))
            dn.append(math.log(dN))
        else:
            t.pop()
    return t, n, dn

N=[10, 100, 1000, 10000, 100000]
plt.figure()
for i in N:
    t, n, dn = decay(i)
    plt.plot(t, n, label = ('N= '+str(i)))
plt.legend()
plt.xlabel('t(s)')
plt.ylabel('log[N(t)]')
plt.savefig('log[N(t)]vsT(s)')
```
